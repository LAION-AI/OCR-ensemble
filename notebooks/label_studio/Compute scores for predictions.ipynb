```python
import sys
sys.path.append('../../')
from ocr_ensemble.evaluation import levenshtein, average_levenshtein, levenshtein_lower
```

```python
import pickle 
import json
```

```python
pred_name = '../../stage1.json'
pred_name = '../../data/label_studio/both_paddle_predictions.pkl'
gt_name = '../../data/label_studio/both_labels.pkl'

with open(gt_name, 'rb') as f:
    gts = pickle.load(f)
if '.pkl' in pred_name:
    with open(pred_name, 'rb') as f:
        preds = pickle.load(f)
else:
    with open(pred_name, 'r') as f:
        preds = json.load(f)

    if isinstance(preds, dict):
        preds = preds['preds']
```

```python
print(average_levenshtein(gts, preds))
print(average_levenshtein(gts, preds, dist=levenshtein_lower))
```
